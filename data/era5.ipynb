# ERA5 Datasets
There are two collections, named below. They are similar, but contain a few different datasets. Here we explore the different datasets, eventually choosing the one to use to train our ML model on.

- weatherbench2:
This dataset collection is provided by the team from weatherbench

- gcp-public-data-arco-era5:
This dataset collection is provided by google research. We use the 'ar' (analysis-ready) parts.

### Dataset types
- *full*: contains full grid res and all 37 levels
- 1959-2022-6h-1440x721: best candidate right now, full grid reduced to 13 levels, and 38 features
- 1959-2023_01_10-wb13-6h-1440x721_with_derived_variables: full grid at 13 levels, with 62 features, many derived
- 

### Grid types
- Full 1440x721 lon/lat grid
- Equiangular conservative with(out) poles at subsampled lon/lat resolutions
- Reduced Gaussian grid (ignored by us)
- Icosahedric mesh made us and GraphCast

# gcp-public-data-arco-era5

```python
era5 = xr.open_zarr(
    "gs://gcp-public-data-arco-era5/ar/1959-2022-full_37-1h-0p25deg-chunk-1.zarr-v2",
    chunks={'time': 48},
    consolidated=True,
)
print("Model surface dataset size {} TiB".format(era5.nbytes/(1024**4)))
era5
```
